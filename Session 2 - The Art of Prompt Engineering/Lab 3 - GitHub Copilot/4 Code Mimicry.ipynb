```python
# The code in this cell is just enough to set up the environment for the next call, which is the actual demo

import pandas as pd
import datetime

modifier = 3
yesterday = datetime.datetime.now() - datetime.timedelta(days=1)
yesterday = yesterday.strftime('%Y-%m-%d')
dfc = pd.read_csv("Sample Data/data.csv")
path1 = "Sample Data/"

def TryGettingValue(row, column):
    try:
        return row[column]
    except:
        return 0
```
