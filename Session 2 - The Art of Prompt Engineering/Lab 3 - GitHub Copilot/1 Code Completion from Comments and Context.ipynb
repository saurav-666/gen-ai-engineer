```python
# there are 4 mental health conditions: Anxiety, Depression, Insomnia, and OCD
# create a pivot table for each condition
# plot the pivot table as a bar chart
# display all 4 bar charts in a single figure  

import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file into a DataFrame
try:
    df = pd.read_csv("Sample Data/data.csv")
except FileNotFoundError:
    print("Error: data.csv not found in Sample Data folder")
    exit()

# Remove NaN values
df = df.dropna()

# List of mental health conditions
conditions = ['Anxiety', 'Depression', 'Insomnia', 'OCD']

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

# Loop through each condition and create a pivot table and bar chart
for i, condition in enumerate(conditions):
    # Group the data by the condition and Music effects and count the occurrences
    grouped_data = df.groupby([condition, 'Music effects']).size().reset_index(name='Count')
    
    # Pivot the grouped data for the bar chart
    try:
        pivot_data = grouped_data.pivot(index=condition, columns='Music effects', values='Count').fillna(0)
    except KeyError:
        print(f"Error: {condition} or Music effects not found in DataFrame")
        continue
    
    # Create the bar chart
    pivot_data.plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'Music Effects vs {condition}')
    axes[i].set_xlabel(f'{condition} Level')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=0)
    
# Adjust layout and display the figure
plt.tight_layout()
plt.show()
    
```
