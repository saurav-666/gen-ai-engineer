import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('data.csv')
data.head()

```python
data.columns
```

```python
# Anxiety is a number from 1 to 10
# Music Effects is either Worsen, No Effect, or Improve

# Let's see how many people have each type of music effect at each anxiety level
# We can use a pivot table for this

# Create a pivot table
pivot_table = data.pivot_table(index='Anxiety', columns='Music effects', values="Timestamp", aggfunc='count', fill_value=0, dropna=True)
pivot_table.head()
```

```python
# plot the pivot table as a bar chart
pivot_table.plot(kind='bar', ylabel='Number of people', xlabel='Anxiety level', title='Music effects on anxiety levels')
```

```python
# there are 4 mental health conditions: Anxiety, Depression, Insomnia, and OCD
# create a pivot table for each condition
# plot the pivot table as a bar chart
# display all 4 bar charts in a single figure

# Anxiety
pivot_table = data.pivot_table(index='Anxiety', columns='Music effects', values="Timestamp", aggfunc='count', fill_value=0, dropna=True)
plt.figure()
pivot_table.plot(kind='bar', ylabel='Number of people', xlabel='Anxiety level', title='Music effects on anxiety levels')

# Depression
pivot_table = data.pivot_table(index='Depression', columns='Music effects', values="Timestamp", aggfunc='count', fill_value=0, dropna=True)
plt.figure()
pivot_table.plot(kind='bar', ylabel='Number of people', xlabel='Depression level', title='Music effects on depression levels')

# Insomnia
pivot_table = data.pivot_table(index='Insomnia', columns='Music effects', values="Timestamp", aggfunc='count', fill_value=0, dropna=True)
plt.figure()
pivot_table.plot(kind='bar', ylabel='Number of people', xlabel='Insomnia level', title='Music effects on insomnia levels')

# OCD
pivot_table = data.pivot_table(index='OCD', columns='Music effects', values="Timestamp", aggfunc='count', fill_value=0, dropna=True)
plt.figure()
pivot_table.plot(kind='bar', ylabel='Number of people', xlabel='OCD level', title='Music effects on OCD levels')

plt.show()
```
