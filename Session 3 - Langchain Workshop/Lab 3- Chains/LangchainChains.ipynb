```python
import sys
sys.path.insert(1, '../../')
import init_creds as creds
 
AZURE_OPENAI_API_KEY = creds.get_api_key()
AZURE_OPENAI_ENDPOINT = creds.get_endpoint()
# print(AZURE_OPENAI_API_KEY)
# print(AZURE_OPENAI_ENDPOINT)
 
if not AZURE_OPENAI_API_KEY:
    raise ValueError("No AZURE_OPENAI_API_KEY set for Azure OpenAI API")
if not AZURE_OPENAI_ENDPOINT:
    raise ValueError("No AZURE_OPENAI_ENDPOINT set for Azure OpenAI API")
```

```python
import os

os.environ["AZURE_OPENAI_API_KEY"] = AZURE_OPENAI_API_KEY
os.environ["AZURE_OPENAI_ENDPOINT"] = AZURE_OPENAI_ENDPOINT
os.environ["AZURE_OPENAI_DEPLOYMENT_NAME"] = "gpt-4o-mini"
os.environ["AZURE_OPENAI_API_VERSION"]="2024-07-01-preview"


from langchain_openai import AzureChatOpenAI

model = AzureChatOpenAI(
    azure_endpoint=os.environ["AZURE_OPENAI_ENDPOINT"],
    azure_deployment=os.environ["AZURE_OPENAI_DEPLOYMENT_NAME"],
    openai_api_version=os.environ["AZURE_OPENAI_API_VERSION"],
)
```

## Generic Chains

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

prompt= PromptTemplate(input_variables=["item"],template="Best {item} in the world")
chain=LLMChain(prompt=prompt,llm=model,verbose=True)
data = ["chocolate", "pizza", "sushi"]
for d in data:
    print(chain.run(d))
```

## Sequential Chain

```python
from langchain.chains import SimpleSequentialChain
prompt= PromptTemplate(input_variables=["n"],template="""Tell me name of {n} famous authors
                        """)
authors_list=LLMChain(llm=model,prompt=prompt, verbose=True)
print("--------authors response-----------")
authors_list.run("3")
```

```python
prompt2=PromptTemplate(input_variables=["country"],template="""Given a list of authors please tell me the {country} where they were born""")
place_chain=LLMChain(llm=model,prompt=prompt2, verbose=True)
print("--------country response-----------")
place_chain.run("country")
```

```python

final_chain=SimpleSequentialChain(chains=[authors_list,place_chain])
print("--------final response-----------")
final_chain.run(input={"n":'5',"country":'country'})
```
